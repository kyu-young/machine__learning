```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split 
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.layers import Dense, Dropout, Activation
from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import SGD
```

```python
mnist = pd.read_csv('/Users/mac/notebook_dir/data/digit-recognizer/numtrain.csv')
```

```python
train_x = mnist.drop('label', axis=1)
train_y = mnist['label']
train_x = np.array(train_x)
train_y = np.array(train_y)
# 정규화
train_x_scale = train_x/255
```

```python
# 계층샘플링
x_train, x_test, y_train, y_test = train_test_split(train_x_scale, train_y, test_size=0.2,
                                                     random_state=42, stratify=train_y)
```

```python
y_train_onehot = tf.one_hot(y_train, depth=10)
y_test_onehot = tf.one_hot(y_test, depth=10)
```

```python
model = Sequential()
model.add(Dense(100, input_shape=[28*28,], kernel_initializer='he_normal'))
model.add(Activation('relu'))
model.add(Dropout(0.2))
model.add(Dense(100, kernel_initializer='he_normal'))
model.add(Activation('relu'))
model.add(Dropout(0.2))
model.add(Dense(10, kernel_initializer='he_normal'))
model.add(Activation('softmax'))
```

```python
model.compile(optimizer = SGD(learning_rate=1e-2), loss = 'categorical_crossentropy', metrics = ['accuracy'])
```

```python
history = model.fit(x_train, y_train_onehot, validation_split = 0.3, epochs = 100, batch_size=100)
```
